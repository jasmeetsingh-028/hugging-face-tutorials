Visual Question Answering

```python
from transformers import BlipForQuestionAnswering
```

```python
model = BlipForQuestionAnswering.from_pretrained(
    "Salesforce/blip-vqa-base"
)
```

```python
from transformers import AutoProcessor

processor = AutoProcessor.from_pretrained(
    "Salesforce/blip-vqa-base"
)
```

```python
from PIL import Image
image = Image.open('Image1.png')
```

```python
question = "How many soccer players are in the picture?"
inputs = processor(image, question, return_tensors = 'pt')
```

```python
out = model.generate(**inputs)
```

```python
print(processor.decode(out[0], skip_special_tokens = True))
```

Clip for zero-shot Image Classification

```python
from transformers import CLIPModel
```

```python
model = CLIPModel.from_pretrained(
    "openai/clip-vit-large-patch14"
)
```

```python
from transformers import AutoProcessor
```

```python
processor = AutoProcessor.from_pretrained(
    "openai/clip-vit-large-patch14"
)
```

```python
labels = ['photo of a soccer player', 'photo of a wrestler']
```

```python
inputs = processor(text = labels,
                   images = image,
                   return_tensors = 'pt',
                   padding = True) #labels lengths are not equal
```

```python
output = model(**inputs)
```

```python
##output logits

output.logits_per_image
```

```python
#logits to probablity scores

probs = output.logits_per_image.softmax(dim = 1)[0]
probs = list(probs)

for idx, label in enumerate(labels):
  print(f'probablity of label {label}: {probs[idx].item():.4f}')
```
