Chatbot using hugging face

```python
from transformers import pipeline
```

```python
chatbot = pipeline(
    task = "conversational",
    model = "facebook/blenderbot-400M-distill"
)  #conversational pipeline
#small model
```

```python
input_text = """
how can i make coffee?
"""

#put this message into conversation object -

from transformers import Conversation

conversation = Conversation(input_text)

model_output = chatbot(conversation)  ##passing input to the model
print(model_output)
```

```python
##continue chatting

print(chatbot(Conversation('what beans should i use?')))  #answer not related to coffee
```

```python
#keeping track of previous prompts and llm responses, helps llm conversing

input_text = """
how can i make coffee?
"""

#put this message into conversation object -

from transformers import Conversation

conversation = Conversation(input_text)

model_output = chatbot(conversation)  ##passing input to the model

conversation.add_message({
    "role":"user",
    "content" :"""
    what beans should i use?
    """
})
print(conversation)
```

Translation and Summarization

Translation

```python
import torch
translator = pipeline(
    task = "translation",
    model = "facebook/nllb-200-distilled-600M",  #meta's no language left behind for lanuage translation (200 different language translations)
    torch_dtype = torch.bfloat16  #compress model without loss in performance
)
```

```python
text = """\
i think we should skip going to the park for the summers.
"""

#language string: https://github.com/facebookresearch/flores/blob/main/flores200/README.md
text_translated = translator(text,
                             src_lang = 'eng_Latn',
                             tgt_lang = 'fra_Latn')

print(text_translated)
```

```python
##deleting models to free momory
import gc
del translator

gc.collect()
```

Summarization

```python
summarizer = pipeline(task = "summarization",
                      model = "facebook/bart-large-cnn",
                      torch_dtype = torch.bfloat16
                      )
```

```python
text_to_summarize = """In the dynamic realm of Natural Language Processing
(NLP), the emergence of models like Llama 2 by Meta AI has ushered in a
new era of possibilities for developers and researchers alike. Llama 2,
a cutting-edge language model, offers the potential for fine-tuning to
specific tasks, making it a versatile tool with a myriad of applications. """
```

```python
summary = summarizer(
    text_to_summarize,
    min_length = 10,
    max_length = 50
)
summary
```

```python
import gc
del summarizer

gc.collect()
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")  #fine tuned by open source community
```

```python
#convert sentences to vectors/embeddings which captures the semantic meaning/information of the sentence
sentences1 = [
    'boys are playing in the park',
    'she is cooking dinner'
]

embeddings1 =  model.encode(sentences1, convert_to_tensor = True)
#embeddings1

sentences2 = [
    'they are playing football.',
    'A man is preparing a meal.'
]

embeddings2 =  model.encode(sentences2, convert_to_tensor = True)

#calculating cosine similarity between the embeddings

from sentence_transformers import util

cosine_sim = util.cos_sim(embeddings1, embeddings2)
print(cosine_sim)  #pairwise sim for every sentence in the first list to every sentence in thew second list

#diagonals are the similarities between 1st and 2nd sentences of both lists
```

```python
for i in range(len(sentences1)):
  print(f'{sentences1[i]} \nand \n{sentences2[i]} \nhave similarity: {cosine_sim[i][i]}')
```
